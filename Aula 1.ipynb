Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
# PASSO A PASSO

# usaremos pyautogui e pandas

# pyautogui -> biblioteca, pacote de codigos, automatiza mouse, teclado e tela do computador.
# pandas -> biblioteca que trabalha base de dados

# pyautogui.press -> uma unica tecla
# pyauotgui.hotkey -> conjunto de teclas
# pyautogui.write -> escreve texto
# pyautogui.PAUSE = 1 -> vai esperar um segundo entre comandos, pode variar o numero
# pyautogui.click -> clica em um local da tela, precisa de uma posição (x,y)
# pyautogui.position -> acha a posição do mouse na tela, utilizado em celula separada  


# pyperclip = copia texto (ctrl+c)
#           = utilizado pois o python não reconhece caracteres espeiais, tipo acento e ?


import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1


# Passo1: Entrar no sistema da empresa (no caso google drive)
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

time.sleep(5)

# Passo2: Navegar no sistema e encontrar a base de vendas (no caso pasta Exportar)
pyautogui.click(x=360, y=299, clicks=2)
time.sleep(2)

# Passo3: Fazer Download da base de Vendas
pyautogui.click(x=357, y=285) #clicar no arquivo
pyautogui.click(x=1072, y=193) #clicar no 3 pontinhos
pyautogui.click(x=888, y=600) #clicar em Fazer Downloads
time.sleep(5) # para esperar o download acabar



 #https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing
 #https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxgausp=sharing

```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# Passo4: Importar base de vendas para Python
import pandas as pd

# se for outro editor(não o Jupyter) tem que instalar e importar: pandas, numpy e openpyxl
# r antes do atalho, para não confundir \, transforma em um texto cru.
tabela = pandas.read_excel(r"C:\Users\autoe\Downloads\Vendas - Dez.xlsx")
display (tabela)

# print exibe a tabela,
# display exibe a tabela formatada pelo python
```

```python
# Passo5: Calcular indicadores da empresa

# Faturamento
faturamento = tabela["Valor Final"].sum()
# tabela[] -> colchetes ele vai entender que é a coluna
# .sum -> soma

# Quantidade produtos vendidos
quantidade = tabela["Quantidade"].sum ()

print (faturamento)
print (quantidade)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# Passo6: Enviar e-mail para diretoria com indicadores da empresa

# abrir aba
pyautogui.hotkey("ctrl", "t")
# entrar link e-mail - https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox
pyperclip.copy("https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
time.sleep(5)

# clicar botão escrever
pyautogui.click(x=97, y=206) #clicar no botão escrever

# preencehr informações do e-mail
pyautogui.write ("autoeletricataquarense@hotmail.com")
pyautogui.press('tab') # para preenceher o e-mail destino
pyautogui.press('tab') # para pular para assunto
time. sleep(3)
# pyautogui.write ("Relatório vendas - Intensivão Python, Aula 1")
# se usar write, não coloca caracteres especiais, tipo acento, então usa:
pyperclip.copy ("Relatório vendas - Intensivão Python, Aula 1")
pyautogui.hotkey ("ctrl","v")

pyautogui.press('tab') # para pular corpo do e-mail

# usa-se 3 aspas """ para escrever texto com varios parágrafos
# f antes do texto, para formatar, isso é, pode-se inserir as variáveis criadas, entre chaves
# formatar numero (moeda, dinheiro)
# :, -> separdor milhar
# .2f -> 2 casas decimais 
texto = f"""
Prezados,


Segue o  relatório de vendas:
Faturamento: R$ {faturamento:,.2f}

Quantidade de produtos vendidos: {quantidade:,}

Qualquer dúvida estou a disposição



Att. Nicole de Rossi


"""
pyperclip.copy (texto)
pyautogui.hotkey ("ctrl","v")

# anexar arquivo
  # clicar anexar
pyautogui.click(x=790, y=951) 
time.sleep(2)
  # escrever o local do arquivo
arquivo = r"C:\Users\autoe\Downloads\Vendas - Dez.xlsx"
pyperclip.copy (arquivo)
pyautogui.hotkey ("ctrl", "v")
  # apertar enter
pyautogui.press ("enter")  
time.sleep(2)

# enviar e-mail
pyautogui.hotkey ("ctrl", "enter") # atalho para enviar

```
